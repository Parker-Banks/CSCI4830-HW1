```python
def forward_euler(N, S, I, R, beta, gamma, dt):
    """
        Calculates a step in a function using the forward euler method.
        
        N: Population size
        S: Number of susceptible individuals
        I: Number of infected individuals
        R: Number of recovered individuals
        beta: Transmission rate constant
        gamma: Per capita recovery rate
        t: Change in time
        
        Returns: New S, I, and R values using forward euler
    """
    SDot = -(beta * S * I) / float(N)
    IDot = (beta * S * I) / float(N) - gamma * I
    RDot = gamma * I
    
    SNext = S + (SDot * dt)
    INext = I + (IDot * dt)
    RNext = R + (RDot * dt)
    
    return SNext, INext, RNext


# Changing
dt = 0.1

# Constants
N = 1000
gamma_arr = [0.5, 0.5, 0.5]
beta_arr = [1, 1.5, 2]
total_time = 50
iterations = int(total_time / dt)

# Calculate values and plot
for ind in range(3):
    S = 999
    I = 1
    R = 0

    SVals, IVals, RVals = [S], [I], [R]
    beta = beta_arr[ind]
    gamma = gamma_arr[ind]
    
    for i in range(iterations - 1):
        S, I, R = forward_euler(N, S, I, R, beta, gamma, dt)
        SVals.append(S)
        IVals.append(I)
        RVals.append(R)
        
    # Create a new figure and axis for each set of beta and gamma values
    fig, ax = plt.subplots()
    
    # Plot S, I, and R for the current set of beta and gamma values
    ax.plot(list(np.linspace(0, total_time, iterations)), SVals, label='Susceptible (Parker)')
    ax.plot(list(np.linspace(0, total_time, iterations)), IVals, label='Infected (Parker)')
    ax.plot(list(np.linspace(0, total_time, iterations)), RVals, label='Recovered (Parker)')
    
    # Add labels and legend for the current set of beta and gamma values
    ax.set_xlabel('Time')
    ax.set_ylabel('Population')
    ax.legend()
    ax.set_title(f'Graph for beta={beta_arr[ind]}, gamma={gamma_arr[ind]}')

    # Show the plot for the current set of beta and gamma values
    plt.show()
```
